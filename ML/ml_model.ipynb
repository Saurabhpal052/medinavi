```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/Adhiraj/Downloads/Original_Dataset.csv")
data.head()
```

```python
columns_to_check = []
for col in data.columns:
    if col != 'Disease':
        columns_to_check.append(col)
```

```python
symptoms1 = data.iloc[:, 1:].values.flatten()
symptoms1 = list(set(symptoms1))
```

```python
for symptom in symptoms1:
    data[symptom] = data.iloc[:, 1:].apply(lambda row: int(symptom in row.values), axis=1)

data1 = data.drop(columns=columns_to_check)
```

```python
data1 = data1.loc[:, data1.columns.notna()]
```

```python
import regex as re
sym = data1.columns
regex = re.compile('_')
sym = [i if regex.search(i) == None else i.replace('_', ' ') for i in sym ]
```

```python
from flashtext import KeywordProcessor
keyword_processor = KeywordProcessor()
keyword_processor.add_keywords_from_list(sym)
text = 'I have itching, joint pain and fatigue'
keyword_processor.extract_keywords(text)
```

```python
def predict_disease(query):
    matched_keyword = keyword_processor.extract_keywords(query)
    if len(matched_keyword) == 0:
        print("No Matches")
    else:
        regex = re.compile(' ')
        processed_keywords = [i if regex.search(i) == None else i.replace(' ', '_') for i in matched_keyword]
        print(processed_keywords)
```

```python
query = 'I have redness of eyes and cough'
predict_disease(query)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import RandomizedSearchCV
from collections import Counter
```

```python
var_mod = ['Disease']
le = LabelEncoder()
for i in var_mod:
    data1[i] = le.fit_transform(data1[i])
```

```python
X = data1.drop(columns="Disease")
y = data1['Disease']
```

```python
def class_algo(model_name,model,independent,dependent):
    model.fit(independent,dependent)
    pred = model.predict(independent)
    accuracy = metrics.accuracy_score(pred,dependent)
    print(model_name,'Accuracy : %s' % '{0:.3%}'.format(accuracy))
```

```python
algorithms = {'Logistic Regression': 
              {"model": LogisticRegression()},
              
              'Decision Tree': 
              {"model": tree.DecisionTreeClassifier()},
              
              'Random Forest': 
              {"model": RandomForestClassifier()},
              
              'SVM':
              {"model": svm.SVC(probability=True)},
              
              'NaiveBayes' :
              {"model": GaussianNB()},
              
              'K-Nearest Neighbors' :
              {"model": KNeighborsClassifier()},
             }
```

```python
for model_name, values in algorithms.items():
    class_algo(model_name,values["model"],X,y)
```

```python
model = LogisticRegression()
model.fit(X,y)
pred = model.predict(X)
accuracy = metrics.accuracy_score(pred,y)
print('LogisticRegression()','Accuracy : %s' % '{0:.3%}'.format(accuracy))
```

```python
test_col = []
for col in data1.columns:
    if col != 'Disease':
        test_col.append(col)


test_data = {}
symptoms = []
predicted = []
def test_input():
    symptoms.clear()
    predicted.clear()
    user_input = input("Enter Symptoms : ")
    matched_keyword = keyword_processor.extract_keywords(user_input)
    if len(matched_keyword) == 0:
        print("No Matches")
        result_df = pd.DataFrame()
    else:
        regex = re.compile(' ')
        processed_keywords = [i if regex.search(i) == None else i.replace(' ', '_') for i in matched_keyword]
        print("Symptoms you have:", processed_keywords)
    
    for column in test_col:
        test_data[column] = 1 if column in symptoms else 0
        test_df = pd.DataFrame(test_data, index=[0])
    predict_disease = model.predict(test_df)
    predict_disease = le.inverse_transform(predict_disease)
    predicted.extend(predict_disease)
    disease_counts = Counter(predicted)
    percentage_per_disease = {disease: (count)  for disease, count in disease_counts.items()}
    result_df = pd.DataFrame({"Disease": list(percentage_per_disease.keys())})
    result_df = result_df.merge(doc_data, on='Disease', how='left')
    result_df = result_df.merge(des_data, on='Disease', how='left')
    result_df = result_df.merge(merged_df, on='Specialist', how='left')
    result_df = result_df.drop_duplicates()
    
    #result_df = result_df.drop('Chances', axis=1)
    result_df = result_df.drop('Description', axis=1)
    return result_df
```
